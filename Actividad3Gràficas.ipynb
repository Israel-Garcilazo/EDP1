A cotinuaciòn el proceso necesario para imnprimir la serie de Fourier asociada a $$ f(x) = exp(2x) $$
En funciòn de cosenos.

```python
from sympy import fourier_series, pi, plot, Piecewise, exp
from sympy.abc import x
import numpy as np

```

```python
f1= exp(2*x)
plot(f1, (x,-2,2),line_color ='blue')
```

```python
s = fourier_series(f1, (x, 0, 1))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p1 = plot(f1, s1, s2, s3, (x, -pi, pi), show=False, legend=True)
```

```python
p1.show()
```
