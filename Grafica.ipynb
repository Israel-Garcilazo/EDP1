 Ejercicio: Graficar la funcion $x^2$

Este còdigo imprime la gràfica de una funciòn de la forma f(x) = y

```python
import matplotlib.pyplot as plt
import numpy as np

def graficar_funcion(funcion, xmin, xmax, titulo=""):
    x = np.linspace(xmin, xmax, 400)
    y = funcion(x)

    plt.plot(x, y)
    plt.title(titulo)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    plt.show()

def funcion_cuadratica(x):
    return x**2

graficar_funcion(funcion_cuadratica, -10, 10, "Gráfica de y = $x^2$")
```
