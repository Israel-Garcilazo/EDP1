### Importar bibliotecas y definir la ED.

Se usará la librería Sympy y se define la ecuación diferencial inicial:

```python
import sympy as sp
sp.init_printing()

x, lam = sp.symbols('x lambda')
f = sp.Function('f')

# Ecuación diferencial original
ode = sp.Eq(f(x).diff(x, 2) + 2*f(x).diff(x) + lam*f(x), 0)
display(sp.Eq(sp.Derivative(f(x), x, 2) + 2*sp.Derivative(f(x), x) + lam*f(x), 0))
```

### Verificación de la forma de Sturm-Liouville

Se calcula que el factor integrante es: \begin{equation} \mu(x) = e^{2x} \end{equation}

por lo que al multiplicarlo por la función original queda como:

```python
mu = sp.exp(2*x)
ode_transformed = sp.simplify(mu * (f(x).diff(x, 2) + 2*f(x).diff(x) + lam*f(x)))
ode_transformed = sp.simplify(ode_transformed)

# Forma de Sturm-Liouville
p = sp.exp(2*x)
r = sp.exp(2*x)
sl_form = sp.Eq(sp.diff(p * f(x).diff(x), x) + lam * r * f(x), 0)
display(sl_form)
```

En donde:

\begin{equation} p(x) = e^{2x} \end{equation}
\begin{equation} q(x) = 0 \end{equation}
\begin{equation} r(x) = e^{2x} \end{equation}


### Solución general de la ecuación

En este paso buscamos dar solución al polinomio característico asociado a la ED, que es de la forma:

\begin{equation} P(r) = r^{2} + 2r + \lambda = 0 \end{equation}

Por lo que los valores quedan como:


```python
# Estas líneas resuelve la ecuacón cuadratica del polinomio característica
r1 = sp.symbols('r1')
char_eq = sp.Eq(r1**2 + 2*r1 + lam, 0)
sols = sp.solve(char_eq, r1)
display(sols)
# Se pretende que λ > 1 y entonces la solución resulta en un oscilador armonico, por lo que estas líneas imprimen la solución del oscilador armonico considerando los valores encontrados.
mu = sp.sqrt(lam - 1)
A, B = sp.symbols('A B')
f_sol = sp.exp(-x) * (A * sp.cos(mu * x) + B * sp.sin(mu * x))
display(f_sol)
```

### Aplicación de condiciones de frontera

Se sustituye sobre la solución anterior las condiciones de frontera y mediante el siguiente código se llega a las eigenfunciones:

```python
cond1 = f_sol.subs(x, 0)
cond2 = f_sol.subs(x, 1)

# Sustitución de la primera condición de frontera: f(0) = 0, que eventualmente implica que A = 0.
cond1_simpl = sp.simplify(cond1)
A_val = sp.solve(sp.Eq(cond1_simpl, 0), A)[0]

# Sustitución de la segunda condición de frontera: f(1) = 0.
f_sol_A0 = f_sol.subs(A, A_val)
cond2_simpl = sp.simplify(f_sol_A0.subs(x, 1))
mu_vals = sp.solve(sp.Eq(cond2_simpl, 0), mu)
display(mu_vals)

# Línea que calcula los valores propios.
n = sp.symbols('n', integer=True, positive=True)
lambda_n = 1 + (n * sp.pi)**2
f_n = sp.exp(-x) * sp.sin(n * sp.pi * x)

display(sp.Eq(sp.Symbol('lambda_n'), lambda_n))
display(sp.Eq(sp.Symbol('f_n(x)'), f_n))
```

### Cálculo de los coeficientes de Fourier para la expansión en serie de Fourier generalizada

En la siguiente línea de códigos se calculan los coeficientes mediante la formula:

En donde u(x) es la función dada inicialmenre y $f_n$ son las eigenfunciones.

```python
u = sp.exp(-x)

# Producto interno con r(x) = e^{2x}
numerador = sp.integrate(u * f_n * r, (x, 0, 1))
denominador = sp.integrate(f_n**2 * r, (x, 0, 1))
c_n = sp.simplify(numerador / denominador)

display(sp.Eq(sp.Symbol('c_n'), c_n))
```

### Cálculo de coeficientes para primeros valores de n

Las siguientes líneas calculan algunos valores de los coeficientes de Fourier y los imprime:

```python
for i in range(1, 6):
    expr = sp.integrate(sp.sin(i * sp.pi * x), (x, 0, 1))
    norm = sp.integrate(sp.sin(i * sp.pi * x)**2, (x, 0, 1))
    coef = expr / norm
    print(f"n = {i}: c_n = {sp.simplify(coef)}")
```

### Cálculo de la serie de Fourier generalizada.

En la siguiente parte del código es importante destacar a la variable "N_terms", pues es esta la que va a permitir modificar el número de términos que el código va a calcular y posteriormente a graficar.

```python
import numpy as np
import matplotlib.pyplot as plt

# Número de términos de la serie
N_terms = 25

# Serie de Fourier generalizada
# Mediante el proceso anterior, buscamos entonces aproximarnos a la función dada u(x)
series = 0
for i in range(1, N_terms + 1):
    coef = sp.integrate(sp.sin(i * sp.pi * x), (x, 0, 1)) / sp.integrate(sp.sin(i * sp.pi * x)**2, (x, 0, 1))
    term = coef * sp.exp(-x) * sp.sin(i * sp.pi * x)
    series += term
# Por ultimo imprimimos la serie con los "n" terminos solicitados:
display(sp.Eq(sp.Symbol('S_approx(x)'), series))
```

### Comparación gráfica entre $u(x)$ y la serie de Fourier

Esta ultima sección del codigo se encarga de graficar y comparar la funcion original con la aproximación dada por la serie de Fourier generalizada.

```python
# Convertimos las funciones simbólicas a funciones numéricas
u_func = sp.lambdify(x, sp.exp(-x), "numpy")
series_func = sp.lambdify(x, series, "numpy")

# Valores de x
x_vals = np.linspace(0, 1, 400)
u_vals = u_func(x_vals)
s_vals = series_func(x_vals)

# Graficamos
plt.figure(figsize=(8, 5))
plt.plot(x_vals, u_vals, label="$u(x) = e^{-x}$", linewidth=2)
plt.plot(x_vals, s_vals, label=f"Aprox. de la serie con N={N_terms}", linestyle='--')
plt.title("Comparación entre $u(x)$ y la serie de Fourier truncada")
plt.xlabel("x")
plt.ylabel("Valor")
plt.grid(True)
plt.legend()
plt.show()
```
