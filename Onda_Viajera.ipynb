```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función f(x') = exp(-((x' - 0.5)^2)/0.01)
def f(xp):
    return np.exp(-((xp - 0.5)**2) / 0.01)

# Parámetros del problema
c = 1.0             # Velocidad de propagación
t = 0.2             # Tiempo en el que observamos la perturbación
x = np.linspace(0, 1, 500)  # Dominio espacial

# (a) Onda viajando a la derecha: x' = x - ct
xp_derecha = x - c * t
onda_derecha = f(xp_derecha)

# (b) Onda viajando a la izquierda: x' = x + ct
xp_izquierda = x + c * t
onda_izquierda = f(xp_izquierda)

# (c) Superposición de ambas ondas
onda_superpuesta = onda_derecha + onda_izquierda

# Graficamos los resultados
plt.figure(figsize=(12, 6))

# Onda a la derecha
plt.subplot(1, 3, 1)
plt.plot(x, onda_derecha, label="Onda a la derecha", color="blue")
plt.title("Onda viajando a la derecha")
plt.xlabel("x")
plt.ylabel("f(x - ct)")
plt.grid()
plt.legend()

# Onda a la izquierda
plt.subplot(1, 3, 2)
plt.plot(x, onda_izquierda, label="Onda a la izquierda", color="green")
plt.title("Onda viajando a la izquierda")
plt.xlabel("x")
plt.ylabel("f(x + ct)")
plt.grid()
plt.legend()

# Superposición
plt.subplot(1, 3, 3)
plt.plot(x, onda_superpuesta, label="Superposición", color="red")
plt.title("Superposición de ondas")
plt.xlabel("x")
plt.ylabel("f(x - ct) + f(x + ct)")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
